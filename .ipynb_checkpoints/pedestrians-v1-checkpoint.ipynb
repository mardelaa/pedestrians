```python
import cv2 # OpenCV biblioteka
import numpy as np
import matplotlib
import imageio
import matplotlib.pyplot as plt
```

```python
kernel = np.ones((3, 3))
```

```python
# fun
def ucitajKlip(indeks):
    return imageio.get_reader('klipovi/video'+str(indeks)+'.mp4',  'ffmpeg')
```

```python
klip = ucitajKlip(1)
frejm = klip.get_data(0)
 
siva = cv2.cvtColor(frejm, cv2.COLOR_RGB2GRAY)
invertovana = 255-siva
 
ret, binarna = cv2.threshold(invertovana, 200, 255, cv2.THRESH_BINARY)
binarna = cv2.erode(binarna, kernel, iterations=1)
 
#trazimo plato
image1, contours, hierarchy = cv2.findContours(binarna, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
 
image1 = frejm.copy()
cv2.drawContours(image1, contours, -1, (255, 0, 0), 1)
 
#plt.imshow(slika)
#plt.show()
 
cnt = max(contours, key=cv2.contourArea)
leftmost = tuple(cnt[cnt[:,:,0].argmin()][0])
rightmost = tuple(cnt[cnt[:,:,0].argmax()][0])
topmost = tuple(cnt[cnt[:,:,1].argmin()][0])
bottommost = tuple(cnt[cnt[:,:,1].argmax()][0])
 
cv2.circle(image1, leftmost, 3, (0, 0, 255), -1)
cv2.circle(image1, rightmost, 3, (255, 0, 0), -1)
cv2.circle(image1, topmost, 3, (0, 255, 0), -1)
cv2.circle(image1, bottommost, 3, (255, 0, 0), -1)
 
a = [int(leftmost[0]+60),int(topmost[1]+50)]
b = [int(topmost[0]-45),int(topmost[1]+40)]
c = [int(topmost[0]-65),int(topmost[1]+100)]
d = [int(leftmost[0]+75),int(topmost[1]+100)]
 
pts = np.array([a,b,c,d], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image1,[pts],True,(0,255,255))
 
a1 = [int(bottommost[0]+70),int(bottommost[1]-60)]
b1 = [int(rightmost[0]-110),int(rightmost[1]-60)]
c1 = [int(rightmost[0]-125),int(rightmost[1]-115)]
d1 = [int(bottommost[0]+75),int(bottommost[1]-115)]
 
pts = np.array([a1,b1,c1,d1], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image1,[pts],True,(0,255,0))
 
#plt.imshow(image1)
#plt.show()
 
f = open("out.txt", "w")
f.write("RA 223-2015, Marko Markovic\n")
f.write("file,count\n")
 
for indeks in range(1,11):
    k = 0
    j = 0
    brojac = 0
    brojac2 = 0
    klip1 = ucitajKlip(indeks)
    for i, im in enumerate(klip1):
        j = j+1
 
        if(j == 25):
            k=k+1            
            zona1 = klip1.get_data(i)
            zona1 = zona1[a[1]+10:d[1],a[0]:b[0]-10]
            #plt.imshow(zona1)
            #plt.show()
            
            #zona1
            ssiva = cv2.cvtColor(zona1, cv2.COLOR_RGB2GRAY)
            sinvertovana = 255-ssiva
 
            sbinarna = cv2.adaptiveThreshold(sinvertovana, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 5)
            sbinarna = cv2.dilate(sbinarna, kernel, iterations=1)
            sbinarna = cv2.erode(sbinarna, kernel, iterations=2)
 
            zona2 = klip1.get_data(i)
            zona2 = zona2[d1[1]:a1[1],d1[0]:c1[0]]
            #plt.imshow(zona2)
            #plt.show()
 
            #zona2
            ssiva1 = cv2.cvtColor(zona2, cv2.COLOR_RGB2GRAY)
            sinvertovana1 = 255-ssiva1
 
            sbinarna1 = cv2.adaptiveThreshold(sinvertovana1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 5)
            sbinarna1 = cv2.dilate(sbinarna1, kernel, iterations=1)
            sbinarna1 = cv2.erode(sbinarna1, kernel, iterations=2)
         
           
            #plt.imshow(sbinarna, 'gray');
            #plt.show()
 
            #plt.imshow(sbinarna1, 'gray');
            #plt.show()
 
 
            ########### Nalazim konture za zonu 1
            img, contours, hierarchy = cv2.findContours(sbinarna1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
 
            contours_barcode = [] #ovde ce biti samo konture koje pripadaju bar-kodu
            for contour in contours: # za svaku konturu
                 center, size, angle = cv2.minAreaRect(contour) # pronadji pravougaonik minimalne povrsine koji ce obuhvatiti celu konturu
                 width, height = size
                 if width > 5 and width < 50 and height > 5 and height < 50: # uslov da kontura pripada bar-kodu
                     contours_barcode.append(contour) # ova kontura pripada bar-kodu
                   
 
            image2 = zona1.copy()
            cv2.drawContours(image2, contours_barcode, -1, (255, 0, 0), 1)
            #plt.imshow(image2)
            #plt.show()
            ########### Nalazim konture za zonu 1 ################################
           
           
            ########### Nalazim konture za zonu 2
            img2, contours2, hierarchy2 = cv2.findContours(sbinarna1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
 
            contours_barcode2 = [] #ovde ce biti samo konture koje pripadaju bar-kodu
            for contour in contours2: # za svaku konturu
                center, size, angle = cv2.minAreaRect(contour) # pronadji pravougaonik minimalne povrsine koji ce obuhvatiti celu konturu
                width, height = size
                if width > 5 and width < 50 and height > 5 and height < 50: # uslov da kontura pripada bar-kodu
                    contours_barcode2.append(contour) # ova kontura pripada bar-kodu
 
            image3 = zona2.copy()
            cv2.drawContours(image3, contours_barcode2, -1, (255, 0, 0), 1)
            #plt.imshow(image3)
            #plt.show()
           
            brojac = brojac + len(contours_barcode2);
            brojac2 = brojac2 + len(contours_barcode) + len(contours_barcode2)
            #print("# %i" % k)
            #print('No. of detected pedestrians in current frame: %d' % len(contours_barcode))
            #print("# %i" % k)
            #print('No. of detected pedestrians in current frame: %d' % len(contours_barcode2))
            j=0
 
    f.write("video%s" % indeks + ".mp4,%s" % brojac + "\n")
           
f.close();
    
    
```
