```python
import numpy as np
import cv2 # OpenCV biblioteka
import matplotlib
import matplotlib.pyplot as plt
import pylab
import imageio
```

```python
# iscrtavanje slika i plotova unutar samog browsera
%matplotlib inline 
# prikaz vecih slika
matplotlib.rcParams['figure.figsize'] = 16,12
kernel = np.ones((5, 5))
```

```python
# fun

def ucitajKlip(indeks):
    putanja = 'klipovi/video'+str(indeks)+'.mp4'
    return imageio.get_reader(putanja,  'ffmpeg')
```

```python
#for indeks in range(1,11)
klip = ucitajKlip(1)
frejm = klip.get_data(0)

siva = cv2.cvtColor(frejm, cv2.COLOR_RGB2GRAY)
invertovana = 255-siva

ret, binarna = cv2.threshold(invertovana, 200, 255, cv2.THRESH_BINARY)
binarna = cv2.erode(binarna, kernel, iterations=1)

#trazimo plato
image1, contours, hierarchy = cv2.findContours(binarna, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

konture = []
for contour in contours: 
    center, size, angle = cv2.minAreaRect(contour)
    width, height = size
    if height > 100: 
        konture.append(contour)

image1 = frejm.copy()
cv2.drawContours(image1, konture, -1, (255, 0, 0), 1)

#plt.imshow(slika)
#plt.show()

cnt = max(konture, key=cv2.contourArea)
leftmost = tuple(cnt[cnt[:,:,0].argmin()][0])
rightmost = tuple(cnt[cnt[:,:,0].argmax()][0])
topmost = tuple(cnt[cnt[:,:,1].argmin()][0])
bottommost = tuple(cnt[cnt[:,:,1].argmax()][0])

cv2.circle(image1, leftmost, 3, (0, 0, 255), -1)
cv2.circle(image1, rightmost, 3, (255, 0, 0), -1)
cv2.circle(image1, topmost, 3, (0, 255, 0), -1)
cv2.circle(image1, bottommost, 3, (255, 0, 0), -1)

pts = np.array([[leftmost[0]+15,topmost[1]+20],[topmost[0]-15,topmost[1]],[topmost[0]-65,topmost[1]+50],[leftmost[0]+65,topmost[1]+65]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image1,[pts],True,(0,255,255))

pts = np.array([[bottommost[0]+70,bottommost[1]-60],[rightmost[0]-110,rightmost[1]-60],[rightmost[0]-125,rightmost[1]-110],[bottommost[0]+75,bottommost[1]-110]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image1,[pts],True,(0,255,0))

plt.imshow(image1)
plt.show()



```
